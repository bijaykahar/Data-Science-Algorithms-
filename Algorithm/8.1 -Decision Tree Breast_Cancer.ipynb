```python
data = pd.read_csv("breast_cancer.csv")
```

```python
del data['Unnamed: 0']
```

```python
X=data.iloc[:,:-1]
y=data.outcome
```

```python
X[0:1].values
```

```python
model= DecisionTreeClassifier(max_depth=5)
model.fit(X,y)
```

```python
m=model.predict(X.values)
```

```python
fig= plt.subplots(nrows = 1,ncols = 1,figsize = (8,8), dpi=300)
tree.plot_tree(model,class_names=['Cancer','Non Cancer'],rounded=True, filled=True,);
```
