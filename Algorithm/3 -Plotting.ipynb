# Generating Muliple Columns

```python
#rcParams['figure.figsize'] = 8,6
fig = plt.figure()
#ax = fig.add_axes([0,0,1,1])
fig, ax = plt.subplots(2,3) # 2,3 means 2 rows and 3 columns - 6 plots
#x[0,0].set_xticks([5,9])
ax[0,0].barh(x,y)
ax[0,1].plot(x,y)
ax[0,2].scatter(x,y)
ax[1,0].plot(x)
ax[1,1].pie(y)
ax[1,2].bar(y,x)
plt.show()
```

```python
cars = pd.read_csv('mtcars.csv')
rcParams['figure.figsize'] = 7,4
mpg = cars.mpg
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.set_xticks(range(32))
ax.set_xticklabels(cars.car_model,rotation=90,fontsize=10)
ax.set_title("Miles per Fallon of cars in mtcars")
ax.set_xlabel('car names')
ax.set_ylabel('miles/gal')
ax.set_ylim([0,45])
ax.annotate('Toyota Corolla',fontsize=15,xy=(19,33.9),xytext=(23,40),arrowprops=dict(facecolor='green'))
ax.bar(range(32),mpg)
```
